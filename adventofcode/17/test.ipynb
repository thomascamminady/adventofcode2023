```python
from rich import print

directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]


sequences = []
for d1 in directions:
    for d2 in directions:
        for d3 in directions:
            sequence = [(0, 0)]
            for _ in [d1, d2, d3]:
                tmp = (_[0] + sequence[-1][0], _[1] + sequence[-1][1])
                if tmp not in sequence:
                    sequence.append(tmp)
            if len(sequence) == 4:
                sequences.append(sequence[::-1])

print(sequences, len(sequences))
```

```python
connections = {}
for sequence in sequences:
    d1, d2, d3, d4 = sequence
    for _ in directions:
        tmp = (_[0] + sequence[-1][0], _[1] + sequence[-1][1])
        if tmp not in sequence:
            tmp_sequence = [d1, d2, d3, d4, tmp]
            straight_line = False
            if (
                len(
                    set(
                        [
                            (
                                tmp_sequence[i][0] - tmp_sequence[i - 1][0],
                                tmp_sequence[i][1] - tmp_sequence[i - 1][1],
                            )
                            for i in [1, 2, 3, 4]
                        ]
                    )
                )
                == 1
            ):
                straight_line = True
            else:
                straight_line = False
            new_sequence = [d2, d3, d4, tmp]
            if not straight_line:
                if tuple(sequence) in connections.keys():
                    connections[tuple(sequence)].append(new_sequence)
                else:
                    connections[tuple(sequence)] = [new_sequence]
print(connections)
```
