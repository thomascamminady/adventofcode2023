```python
with open("old.txt", "r") as f:
    c = f.read()
old_numbers = [int(_.split(" ")[1]) for _ in c.splitlines()]
```

```python
with open("new.txt", "r") as f:
    c = f.read()
new_numbers = [int(_.split(" ")[1]) for _ in c.splitlines()]
```

```python
with open(
    "/Users/thomascamminady/Repos/adventofcode2023/adventofcode/riddle_inputs/12.txt",
    "r",
) as f:
    lines = f.read()
lines = lines.splitlines()
```

```python
for i, (oldi, newi) in enumerate(zip(old_numbers, new_numbers)):
    if oldi != newi:
        print(
            oldi - newi,
            len([_ for _ in lines[i].split(" ")[1].split(",")]),
            i,
            oldi,
            newi,
            lines[i],
        )
```
